```python
from datetime import datetime
from pprint import pprint
from time import sleep
import csv
import locale
import numpy as np
import pandas as pd
import requests
```

```python
#Creando Dataframe

archivo = "extraccion-clima-malaga.csv"

df_malaga = pd.read_csv(archivo, header = None)
```

```python
archivo = "extraccion-clima-madrid.csv"

df_madrid = pd.read_csv(archivo, header = None)
```

```python
archivo = "extraccion-clima-coruña.csv"

df_coruña = pd.read_csv(archivo, header = None)
```

```python
#Unir los dataframes

df_total = pd.concat([df_malaga, df_madrid, df_coruña])
```

```python
def limpiar_data(df):
    #Nombrar cabeceras
    columnas = ["Fecha", "Temp Max", "Temp Min", "Lluvia", "Tiempo", "Viento", "Ubicacion"]
    df.columns = columnas
    
    #Set Español para que datetime entienda los meses en español
    #locale.setlocale(locale.LC_TIME, 'es_ES.UTF-8')
    #Cambiar columna de fecha a tipo datetime
    df["Fecha"] = pd.to_datetime(df["Fecha"], format = "%d de %B %Y")
    
    #Modificacion para columna de lluvia
    #Eliminar mm
    df["Lluvia"] = df["Lluvia"].replace({"mm":""}, regex=True)
    #Eliminar espacios blancos
    df["Lluvia"] = df["Lluvia"].str.strip()
    #Reemplaza comas por puntos
    df["Lluvia"] = df["Lluvia"].replace({",":"."}, regex = True)
    #Convierte a float la columna lluvia
    df["Lluvia"] = df["Lluvia"].astype(float)
    
    #Modificacion para columna de vientos
    df["Viento"] = df["Viento"].replace({"Km/h":""}, regex=True)
    #Eliminar espacios blancos
    df["Viento"] = df["Viento"].str.strip()
    #Convierte a float la columna lluvia
    df["Viento"] = df["Viento"].astype(int)
    
```

```python
limpiar_data(df_total)
```
